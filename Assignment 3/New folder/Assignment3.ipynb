```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from scipy import stats
```

```python
df = pd.read_csv('gapminder-FiveYearData.csv')
print(df)
```

```python
print(df.duplicated().value_counts())
```

```python
df2=df[['country','year','lifeExp']] 
print(df2)
```

```python
 #create a pivot table as instructed
df3=pd.pivot_table(df2, index='country', columns='year', values='lifeExp') 
print("Pivote_table:\n",df3)
```

```python
#assigning size to fig of heatma[
plt.figure(figsize=(8,8))

#creating heat map and providing color and margins
sns.heatmap(df3,linewidths=.8,fmt="g",cmap="plasma").get_figure().savefig('heatmap.png')

#A title for the image
plt.title("FiveYearData")
```
