```python
import pandas as pd
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds1 = randn(1000)
ds2 = randn(100)
```

```python
plt.hist(ds1)
plt.savefig('image1.png')
plt.show()
```

```python
plt.hist(ds2)
plt.savefig('image2.png')
plt.show()
```

```python
#since we using more than 1 block we have to normalize the function
#alpha- transparency factor(to make the color transparent)
plt.hist(ds1, normed=True, color='green', bins=15, alpha=0.5)
plt.hist(ds2, normed=True, bins=15, alpha=0.5)
plt.savefig('image3.png')
plt.show()
```

```python
#Seaborn
ds3 = randn(2000)
ds4 = randn(2000)
```

```python
sns_plot = sns.jointplot(ds3,ds4)
sns_plot.savefig('image4.png')
```

```python

sns_plot2 = sns.jointplot(ds3,ds4,kind='hex')
sns_plot2.savefig('image5.png')
```
